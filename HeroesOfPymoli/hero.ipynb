```python
import pandas as pd
import numpy as np
import json
```

```python
hero_file="purchase_data.json"
info=pd.read_json(hero_file)
info.head()
```

```python
number_of_players=info['SN'].nunique()

Total_Players= pd.DataFrame({"Total Players":number_of_players}, index=[0])
Total_Players
```

```python
## Purchasing Analysis(Total)
num_of_uniq_items=info["Item Name"].nunique()
avg_price=info['Price'].mean()
num_of_purch=info['Item ID'].count()
tot_reven=info['Price'].sum()

Purchases_tool_table= pd.DataFrame({"Number of Unique Items":num_of_uniq_items, "Average Purchase Price":avg_price,
                                    "Total Number of Purchases":num_of_purch,"Total Revenue":tot_reven}, index=[0])
Purchases_tool_table
```

```python
info["Gender"].unique()
```

```python
## Gender Demographics

male_filter=info.loc[info['Gender']=='Male']
Female_filter=info.loc[info['Gender']=='Female']
Other_filter=info.loc[info['Gender']=='Other / Non-Disclosed']
##counts
male_count=male_filter['SN'].nunique()
female_count=Female_filter['SN'].nunique()
other_count=Other_filter['SN'].nunique()
## Percentages
male_percentage=(male_count/number_of_players)*100
female_percentage=(female_count/number_of_players)*100
other_percentage=(other_count/number_of_players)*100
```

```python
info.head()
```

```python
pd.DataFrame(info.groupby("Gender").nunique()['SN'].rename('count'))
```

```python
d=pd.DataFrame({"Percentage":[male_percentage,female_percentage,other_percentage],"Total Count":[male_count,female_count,other_count]}, index=['Male','Female','Other / Non-Disclosed'])
d
```

```python
## Purchase Analysis(gender)
gender_group= info.groupby("Gender")

purch_count=gender_group['Item ID'].count()
avg_price=gender_group['Price'].mean()
tot_value=gender_group['Price'].sum()
norm= tot_value/d['Total Count']

Purchasing_Analysis_Gender_table= pd.DataFrame({"Purchase Count":purch_count,"Average Purchase Price":avg_price,
                                               "Total Purchase Value":tot_value,"Normalized Totals":norm})
```

```python
Purchasing_Analysis_Gender_table
```

```python
print(info['Age'].min())
print(info['Age'].max())
```

```python
bins=[0,9,14,19,24,29,34,39,50]
group_names=["<10",'10-14','15-19','20-24','25-29','30-34','35-39','40+']
age_dis=info
age_dis['Age Distribution']=pd.cut(info["Age"], bins, labels=group_names)
age_group=age_dis.groupby('Age Distribution')
age_count=age_group['Age'].nunique()
perc=(age_count/info['Age'].count())*100
d=pd.DataFrame({'Percentage of Players':perc,"Total Count":age_count})
d
```

```python
purch_count=age_group['Item ID'].count()
avg_price=age_group['Price'].mean()
tot_value=age_group['Price'].sum()
norm= tot_value/d['Total Count']
Purchasing_Analysis_Age_table= pd.DataFrame({"Purchase Count":purch_count,"Average Purchase Price":avg_price,
                                               "Total Purchase Value":tot_value,"Normalized Totals":norm})
```

```python
Purchasing_Analysis_Age_table
```

```python
sn_group=info.groupby('SN')
sn_purh_count=sn_group['SN'].count()
sn_avg_pric=sn_group['Price'].mean()
sn_tot_pric=sn_group['Price'].sum()
sn_table=pd.DataFrame({"Purchase Count":sn_purh_count,"Item Price":sn_avg_pric,
                       "Total Purchase Value":sn_tot_pric})
```

```python

high_to_low_sort=sn_table.sort_values("Total Purchase Value", ascending=False)
## Retrieving the top 5 record
highest_passing_scores=high_to_low_sort.iloc[0:5,:]
##viewing the results
highest_passing_scores
```
